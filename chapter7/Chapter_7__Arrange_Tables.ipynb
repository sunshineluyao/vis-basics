## Scatterplots: Quantitative Values

Scatterplots utilize two quantitative attributes, leveraging horizontal and vertical spatial channels. They are particularly effective for identifying:

- **Trends**: Overall patterns or changes in data.
- **Outliers**: Data points that deviate from the general pattern.
- **Correlations**: Relationships between two variables.

**Guiding Question 3:** How scatterplots encode quantitative relationships.

**Use Case:** Investigating correlations between income and life expectancy.


### correlation

````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for Correlation Analysis
x_corr = np.random.normal(50, 10, 100)
y_corr = x_corr + np.random.normal(0, 5, 100)

plt.scatter(x_corr, y_corr)
plt.title('Correlation Analysis: Income vs Life Expectancy')
plt.xlabel('Income')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
````

### cluster identification

````python
# Sample data for Cluster Identification
x1_cluster = np.random.normal(40, 5, 50)
y1_cluster = x1_cluster + np.random.normal(0, 5, 50)

x2_cluster = np.random.normal(60, 5, 50)
y2_cluster = x2_cluster + np.random.normal(0, 5, 50)

plt.scatter(x1_cluster, y1_cluster, label='Cluster 1')
plt.scatter(x2_cluster, y2_cluster, label='Cluster 2')
plt.title('Cluster Identification: Subgroups in Data')
plt.xlabel('Income')
plt.ylabel('Life Expectancy')
plt.legend()
plt.grid(True)
plt.show()
````

### outlier

````python
# Sample data with Outliers
x_outlier = np.random.normal(50, 10, 98)
y_outlier = x_outlier + np.random.normal(0, 5, 98)

# Add outliers
x_outlier = np.append(x_outlier, [20, 100])
y_outlier = np.append(y_outlier, [80, 30])

plt.scatter(x_outlier, y_outlier)
plt.title('Outlier Detection')
plt.xlabel('Income')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
````

### Trend Visualization

````python
# Sample data for Trend Visualization
x_trend = np.linspace(10, 100, 100)
y_trend = 3 * np.sin(0.1 * x_trend) + np.random.normal(0, 1, 100)

plt.scatter(x_trend, y_trend)
plt.title('Trend Visualization: Non-linear Relationship')
plt.xlabel('Income')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
````

#### **Correlation Analysis**
- **Key**: Relationship Type  
- **Value**: Positive, Negative, or None (interpreted through visual inspection of trend)

---

#### **Cluster Identification**
- **Key**: Cluster Labels  
- **Value**: Distinct subgroups in data (e.g., Cluster 1, Cluster 2)

---

#### **Outlier Detection**
- **Key**: Outliers  
- **Value**: Points significantly deviating from the central pattern (e.g., [20, 80] or [100, 30])

---

#### **Trend Visualization**
- **Key**: Trend Pattern  
- **Value**: Linear, Non-linear, Exponential, etc., as identified visually by point alignment in scatterplot

## Parallel Coordinates: Multivariate Analysis

Parallel coordinates visualize multiple quantitative dimensions by:

- Representing each attribute on parallel axes.
- Encoding data points as polylines crossing these axes.

**Guiding Question 3:** How parallel coordinates handle high-dimensional data.

**Use Case:** Analyzing customer behavior across various metrics.

## **Exploring Spatial Axis Orientations**





#### **1. Rectilinear Layouts**
- **Use Case**: Ideal for comparing trends over time or analyzing relationships between two variables.
- **Key Features**: Uses perpendicular axes; best for linear relationships and quantitative comparisons.

#### **2. Radial Layouts**
- **Use Case**: Suitable for visualizing cyclic data, such as seasonal trends or periodic data.
- **Key Features**: Uses polar coordinates, emphasizing the cyclical nature of relationships. Visually striking but less precise.

#### **3. Parallel Layouts**
- **Use Case**: Useful for exploring relationships across multiple dimensions or attributes.
- **Key Features**: Enables multi-dimensional analysis, powerful for identifying trends and correlations but can be challenging to interpret.

---

### **Python Code for Visualizations**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import parallel_coordinates

# Rectilinear Layout: Line Chart Example
x_rect = np.linspace(2000, 2024, 25)
y_rect = np.random.normal(50, 10, 25)

plt.figure(figsize=(8, 5))
plt.plot(x_rect, y_rect, marker='o', linestyle='-', color='blue')
plt.title('Rectilinear Layout: Linear Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Radial Layout: Polar Plot Example
theta = np.linspace(0, 2 * np.pi, 30)
r = 5 + 2 * np.sin(3 * theta)

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, r, linestyle='-', marker='o')
ax.set_title('Radial Layout: Cyclical Data Representation', va='bottom')
plt.show()

# Parallel Layout: Multi-Dimensional Data Example
data_parallel = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Attribute1': [5, 7, 6, 8],
    'Attribute2': [15, 10, 13, 14],
    'Attribute3': [20, 23, 19, 21]
})

plt.figure(figsize=(10, 6))
parallel_coordinates(data_parallel, class_column='Category', colormap='Dark2')
plt.title('Parallel Layout: Multi-Dimensional Analysis')
plt.xlabel('Attributes')
plt.grid(True)
plt.show()
```

### **Conclusion**
- **Rectilinear Layouts**: Simple and intuitive for most time-series or linear relationships.
- **Radial Layouts**: Highlight cyclic patterns effectively, ideal for periodic data.
- **Parallel Layouts**: Powerful for high-dimensional data, uncovering correlations across multiple attributes.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import parallel_coordinates

# Rectilinear Layout: Line Chart Example
x_rect = np.linspace(2000, 2024, 25)
y_rect = np.random.normal(50, 10, 25)

plt.figure(figsize=(8, 5))
plt.plot(x_rect, y_rect, marker='o', linestyle='-', color='blue')
plt.title('Rectilinear Layout: Linear Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Radial Layout: Polar Plot Example
theta = np.linspace(0, 2 * np.pi, 30)
r = 5 + 2 * np.sin(3 * theta)

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, r, linestyle='-', marker='o')
ax.set_title('Radial Layout: Cyclical Data Representation', va='bottom')
plt.show()

# Parallel Layout: Multi-Dimensional Data Example
data_parallel = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Attribute1': [5, 7, 6, 8],
    'Attribute2': [15, 10, 13, 14],
    'Attribute3': [20, 23, 19, 21]
})

plt.figure(figsize=(10, 6))
parallel_coordinates(data_parallel, class_column='Category', colormap='Dark2')
plt.title('Parallel Layout: Multi-Dimensional Analysis')
plt.xlabel('Attributes')
plt.grid(True)
plt.show()
````

## **Idioms: Categorical Data Visualization**




When dealing with categorical data, several visualization idioms come to the forefront, each offering unique advantages for different data structures and analytical goals.

#### **Bar Charts**
- **Data Structure**: One key, one value
- **Key Features**: Easy comparison, clear ranking

Bar charts are excellent for comparing values across distinct categories. They are simple to interpret and ideal for presenting ranked data.

#### **Stacked Bar Charts**
- **Data Structure**: Two keys, one value
- **Key Features**: Part-to-whole relationships, category composition

Stacked bar charts provide additional insights by showing the distribution of parts within each category, helping analyze the composition.

#### **Heatmaps**
- **Data Structure**: Two keys, one value
- **Key Features**: High information density, pattern recognition

Heatmaps are compact, leveraging color gradients to highlight patterns or correlations across two categorical dimensions.

#### **Parallel Coordinates**
- **Data Structure**: Multiple quantitative attributes
- **Key Features**: Multi-dimensional analysis, trend identification

Parallel coordinates allow for the visualization of high-dimensional data, helping identify trends and correlations across multiple variables.

---

### **Python Code for Visualization**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.plotting import parallel_coordinates

# Sample data generation
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart: Easy Comparison')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Stacked Bar Chart
data_stacked = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Subgroup1': [3, 5, 2, 6],
    'Subgroup2': [2, 2, 1, 2]
})
data_stacked.set_index('Category').plot(kind='bar', stacked=True, colormap='viridis', figsize=(8, 5))
plt.title('Stacked Bar Chart: Part-to-Whole Relationship')
plt.ylabel('Value')
plt.xlabel('Category')
plt.grid(True)
plt.show()

# Heatmap
heatmap_data = np.random.rand(5, 5)
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', xticklabels=categories, yticklabels=categories)
plt.title('Heatmap: High Information Density')
plt.show()

# Parallel Coordinates
data_parallel = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Attribute1': [10, 15, 14, 17],
    'Attribute2': [20, 24, 22, 30],
    'Attribute3': [5, 9, 8, 12]
})

plt.figure(figsize=(10, 6))
parallel_coordinates(data_parallel, class_column='Category', colormap='Dark2')
plt.title('Parallel Coordinates: Multi-Dimensional Analysis')
plt.xlabel('Attributes')
plt.grid(True)
plt.show()
```

### **Conclusion**
- **Bar Charts**: Best for single-value comparisons.
- **Stacked Bar Charts**: Provide composition insights.
- **Heatmaps**: Use color for dense data insights.
- **Parallel Coordinates**: Unravel multi-dimensional relationships visually.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.plotting import parallel_coordinates

# Sample data generation
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart: Easy Comparison')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Stacked Bar Chart
data_stacked = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Subgroup1': [3, 5, 2, 6],
    'Subgroup2': [2, 2, 1, 2]
})
data_stacked.set_index('Category').plot(kind='bar', stacked=True, colormap='viridis', figsize=(8, 5))
plt.title('Stacked Bar Chart: Part-to-Whole Relationship')
plt.ylabel('Value')
plt.xlabel('Category')
plt.grid(True)
plt.show()

# Heatmap
heatmap_data = np.random.rand(5, 5)
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', xticklabels=categories, yticklabels=categories)
plt.title('Heatmap: High Information Density')
plt.show()

# Parallel Coordinates
data_parallel = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Attribute1': [10, 15, 14, 17],
    'Attribute2': [20, 24, 22, 30],
    'Attribute3': [5, 9, 8, 12]
})

plt.figure(figsize=(10, 6))
parallel_coordinates(data_parallel, class_column='Category', colormap='Dark2')
plt.title('Parallel Coordinates: Multi-Dimensional Analysis')
plt.xlabel('Attributes')
plt.grid(True)
plt.show()
````


### **Dense and Space-Filling Layouts: Maximizing Information Display**

#### **1. Dense Layouts**
- **Use Case**: Ideal for analyzing large datasets to uncover patterns, trends, or outliers.
- **Key Features**: Uses pixel-based visualization techniques to maximize the number of displayed data points, reducing individual mark sizes.

#### **2. Space-Filling Layouts**
- **Use Case**: Excellent for visualizing hierarchical data, such as organizational structures or category breakdowns.
- **Key Features**: Uses nested shapes (e.g., treemaps or sunburst diagrams) to efficiently represent part-to-whole relationships.

### **Comparative Analysis of Visualization Idioms**

#### **1. Scatterplot**
- **Use Case**: Explore relationships and detect outliers between two quantitative variables.
- **Task**: Correlation and outlier detection.
- **Scale**: Hundreds of items.

#### **2. Bar Chart**
- **Use Case**: Compare categorical data (e.g., sales by region).
- **Task**: Comparison and ranking.
- **Scale**: Dozens to hundreds of items.

#### **3. Heatmap**
- **Use Case**: Identify patterns in a matrix of data (e.g., correlation matrix).
- **Task**: Pattern recognition.
- **Scale**: Thousands of items.

#### **4. Parallel Coordinates**
- **Use Case**: Analyze multiple quantitative attributes (e.g., customer segmentation).
- **Task**: Multi-dimensional analysis.
- **Scale**: Dozens of attributes.

#### **5. Pie Chart**
- **Use Case**: Show part-to-whole relationships (e.g., market share distribution).
- **Task**: Part-to-whole analysis.
- **Scale**: One dozen categories.

### **Python Code for Visualizations**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.plotting import parallel_coordinates

# 1. Scatterplot
x_scatter = np.random.normal(50, 10, 100)
y_scatter = x_scatter + np.random.normal(0, 5, 100)

plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, color='blue', alpha=0.6)
plt.title('Scatterplot: Correlation & Outliers')
plt.xlabel('X-axis: Variable 1')
plt.ylabel('Y-axis: Variable 2')
plt.grid(True)
plt.show()

# 2. Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [30, 45, 25, 50]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='orange')
plt.title('Bar Chart: Category Comparison')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# 3. Heatmap
heatmap_data = np.random.rand(10, 10)

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=False, cmap='coolwarm')
plt.title('Heatmap: Pattern Recognition')
plt.show()

# 4. Parallel Coordinates
data_parallel = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Attribute1': [1, 3, 5, 7],
    'Attribute2': [2, 4, 6, 8],
    'Attribute3': [10, 12, 14, 16]
})

plt.figure(figsize=(10, 6))
parallel_coordinates(data_parallel, class_column='Category', colormap='Dark2')
plt.title('Parallel Coordinates: Multi-Dimensional Analysis')
plt.grid(True)
plt.show()

# 5. Pie Chart
sizes = [15, 30, 45, 10]
labels = ['Segment A', 'Segment B', 'Segment C', 'Segment D']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Pie Chart: Part-to-Whole Relationship')
plt.axis('equal')
plt.show()
```

### **Conclusion**
- **Scatterplot**: Visualizes correlation and identifies outliers in data.
- **Bar Chart**: Compares values across categories.
- **Heatmap**: Reveals dense patterns in matrix data.
- **Parallel Coordinates**: Provides a holistic view of multi-dimensional datasets.
- **Pie Chart**: Demonstrates proportions of a whole effectively.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.plotting import parallel_coordinates

# 1. Scatterplot
x_scatter = np.random.normal(50, 10, 100)
y_scatter = x_scatter + np.random.normal(0, 5, 100)

plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, color='blue', alpha=0.6)
plt.title('Scatterplot: Correlation & Outliers')
plt.xlabel('X-axis: Variable 1')
plt.ylabel('Y-axis: Variable 2')
plt.grid(True)
plt.show()

# 2. Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [30, 45, 25, 50]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='orange')
plt.title('Bar Chart: Category Comparison')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# 3. Heatmap
heatmap_data = np.random.rand(10, 10)

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=False, cmap='coolwarm')
plt.title('Heatmap: Pattern Recognition')
plt.show()

# 4. Parallel Coordinates
data_parallel = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Attribute1': [1, 3, 5, 7],
    'Attribute2': [2, 4, 6, 8],
    'Attribute3': [10, 12, 14, 16]
})

plt.figure(figsize=(10, 6))
parallel_coordinates(data_parallel, class_column='Category', colormap='Dark2')
plt.title('Parallel Coordinates: Multi-Dimensional Analysis')
plt.grid(True)
plt.show()

# 5. Pie Chart
sizes = [15, 30, 45, 10]
labels = ['Segment A', 'Segment B', 'Segment C', 'Segment D']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Pie Chart: Part-to-Whole Relationship')
plt.axis('equal')
plt.show()
````
