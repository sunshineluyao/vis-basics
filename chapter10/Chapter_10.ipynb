### Key Considerations When Using Color Channels
- Balance between hue, luminance, and saturation is essential for readability and accuracy.
- **Hue**: Useful for categorization.
- **Luminance**: Provides contrast and ensures text readability.
- **Saturation**: Indicates intensity; useful for highlighting.


````python
import matplotlib.pyplot as plt
import numpy as np

# Create an example demonstrating hue, luminance, and saturation
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, color='blue', label='High Luminance')
plt.plot(x, y2, color='darkblue', label='Low Luminance')
plt.fill_between(x, y1, y2, color='lightblue', alpha=0.5, label='Low Saturation')

plt.title("Hue, Luminance, and Saturation in Visualization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
````

### Advantages and Limitations of Color Spaces
- **RGB**: Computationally efficient but poor for perceptual tasks.
- **HSL/HSV**: Intuitive for designers; good for conceptual mappings.
- **L*a*b***: Perceptually accurate but computationally intensive.

from matplotlib.colors import rgb_to_hsv, hsv_to_rgb

````python
def display_colors(rgb_color, space='RGB'):
    # Prepare RGB, HSV, and L*a*b*
    hsv_color = rgb_to_hsv([[rgb_color]])[0][0]
    lab_color = rgb_to_lab(rgb_color)

    # Convert back for visualization
    rgb_from_hsv = hsv_to_rgb([[hsv_color]])[0][0]
    rgb_from_lab = lab_to_rgb(lab_color)

    fig, ax = plt.subplots(1, 3, figsize=(15, 6))

    # Original RGB
    ax[0].imshow([[rgb_color]])
    ax[0].set_title(f"Original ({space})\n\nRGB: {np.round(rgb_color, 3)}", fontsize=10)

    # HSV Conversion
    ax[1].imshow([[rgb_from_hsv]])
    ax[1].set_title(f"HSV Conversion\n\nHSV: {np.round(hsv_color, 3)}", fontsize=10)

    # L*a*b* Conversion
    ax[2].imshow([[rgb_from_lab]])
    ax[2].set_title(f"L*a*b* Conversion\n\nL*a*b*: {np.round(lab_color, 3)}", fontsize=10)

    # Remove axis ticks
    for a in ax:
        a.axis('off')

    # Adjust spacing to prevent text overlap
    plt.tight_layout()
    plt.show()

# Interactive widget
def update_color(r=0, g=0, b=0, space='RGB'):
    rgb_color = [r / 255, g / 255, b / 255]  # Normalize RGB
    display_colors(rgb_color, space)

# Sliders for RGB values
r_slider = widgets.IntSlider(value=255, min=0, max=255, step=1, description='Red')
g_slider = widgets.IntSlider(value=255, min=0, max=255, step=1, description='Green')
b_slider = widgets.IntSlider(value=255, min=0, max=255, step=1, description='Blue')
space_dropdown = widgets.Dropdown(options=['RGB', 'HSV', 'L*a*b*'], description='Color Space')

# Link sliders to update function
ui = widgets.VBox([space_dropdown, r_slider, g_slider, b_slider])
out = widgets.interactive_output(update_color, {'r': r_slider, 'g': g_slider, 'b': b_slider, 'space': space_dropdown})

# Display the dashboard
display(ui, out)
````

### How Colormaps Support Data Visualization
- **Categorical**: Useful for group distinctions.
- **Sequential**: Represents ordered data.
- **Diverging**: Highlights critical values with two diverging tones.


````python
import matplotlib.pyplot as plt
import numpy as np

# Sequential and Diverging Colormaps Example
data = np.random.rand(10, 10)

plt.figure(figsize=(12, 6))

# Sequential Colormap
plt.subplot(1, 2, 1)
plt.imshow(data, cmap='viridis')
plt.colorbar()
plt.title("Sequential Colormap (Viridis)")

# Diverging Colormap
plt.subplot(1, 2, 2)
plt.imshow(data - 0.5, cmap='coolwarm')
plt.colorbar()
plt.title("Diverging Colormap (Coolwarm)")

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# Generate sample data
def generate_data(data_type):
    np.random.seed(42)
    if data_type == "Categorical":
        # Categorical data: group distinctions
        data = np.random.randint(0, 5, (10, 10))
        colormap = "tab10"
        title = "Categorical Colormap (Tab10)"
    elif data_type == "Sequential":
        # Sequential data: ordered data (e.g., temperature)
        data = np.linspace(0, 1, 100).reshape(10, 10)
        colormap = "viridis"
        title = "Sequential Colormap (Viridis)"
    elif data_type == "Diverging":
        # Diverging data: values centered around a critical point
        data = np.random.randn(10, 10)
        colormap = "coolwarm"
        title = "Diverging Colormap (Coolwarm)"
    else:
        raise ValueError("Invalid data type selected!")
    return data, colormap, title

# Plot the data using the specified colormap
def plot_colormap(data_type):
    clear_output(wait=True)
    data, colormap, title = generate_data(data_type)
    plt.figure(figsize=(6, 5))
    plt.imshow(data, cmap=colormap)
    plt.colorbar()
    plt.title(title)
    plt.axis("off")
    plt.show()

# Dropdown widget to select colormap type
data_type_selector = widgets.Dropdown(
    options=["Categorical", "Sequential", "Diverging"],
    description="Data Type:",
    value="Categorical"
)

# Link dropdown to plot function
interactive_plot = widgets.interactive_output(plot_colormap, {"data_type": data_type_selector})

# Display the interactive widget
display(widgets.VBox([data_type_selector, interactive_plot]))
````

````python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to generate bar chart for a dataset
def generate_barchart(use_safe_colors, adjust_luminance):
    clear_output(wait=True)

    # Sample data for the bar chart
    categories = ["Category A", "Category B", "Category C"]
    values = [15, 25, 10]

    # Colors for the chart
    if use_safe_colors:
        # Colorblind-safe: blue-orange scheme
        colors = ["blue", "orange", "purple"]
    else:
        # Problematic: red-green scheme
        colors = ["red", "green", "brown"]

    # Adjust luminance (if enabled)
    if adjust_luminance:
        colors = ["#4477AA", "#EE6677", "#AA4499"]  # High-contrast, colorblind-safe palette

    # Plot the bar chart
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.bar(categories, values, color=colors)

    # Add labels and title
    ax.set_title("Colorblind-Safe Bar Chart" if use_safe_colors else "Problematic Red-Green Bar Chart")
    ax.set_ylabel("Values")
    ax.set_ylim(0, 30)
    plt.show()

# Function to simulate colorblind-safe design
def simulate_design():
    print("Testing colors for accessibility:")
    print("- Avoiding red-green combinations.")
    print("- Testing designs for colorblind safety.")
    print("- Adjusting luminance for readability.")

# Widgets for interactivity
safe_colors_toggle = widgets.Checkbox(
    value=False, description="Use Colorblind-Safe Colors", indent=False
)
luminance_toggle = widgets.Checkbox(
    value=False, description="Adjust Luminance", indent=False
)
simulate_button = widgets.Button(description="Simulate Design")

# Event handlers
def update_chart(change):
    generate_barchart(safe_colors_toggle.value, luminance_toggle.value)

def on_simulate_click(b):
    simulate_design()

# Link widgets to functionality
safe_colors_toggle.observe(update_chart, names="value")
luminance_toggle.observe(update_chart, names="value")
simulate_button.on_click(on_simulate_click)

# Initial output
generate_barchart(False, False)

# Display the interactive widgets
display(widgets.VBox([safe_colors_toggle, luminance_toggle, simulate_button]))
````

````python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to demonstrate red-green confusion and low luminance contrast
def demonstrate_challenges(confusion_type="Red-Green"):
    clear_output(wait=True)  # Clear previous output
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    if confusion_type == "Red-Green":
        # Problematic red-green combination
        colors = ["red", "green"]
        labels = ["Red", "Green"]
        ax[0].bar([1, 2], [10, 15], color=colors)
        ax[0].set_xticks([1, 2])
        ax[0].set_xticklabels(labels)
        ax[0].set_title("Red-Green Confusion (Problematic)")

        # Improved blue-orange combination
        improved_colors = ["blue", "orange"]
        ax[1].bar([1, 2], [10, 15], color=improved_colors)
        ax[1].set_xticks([1, 2])
        ax[1].set_xticklabels(["Blue", "Orange"])
        ax[1].set_title("Improved: Blue-Orange Pair (Colorblind Safe)")

    elif confusion_type == "Low-Luminance":
        # Low contrast (yellow-white)
        colors = ["yellow", "white"]
        labels = ["Yellow", "White"]
        ax[0].bar([1, 2], [10, 15], color=colors)
        ax[0].set_xticks([1, 2])
        ax[0].set_xticklabels(labels)
        ax[0].set_title("Low Luminance Contrast")

        # High contrast (blue-white)
        improved_colors = ["blue", "white"]
        ax[1].bar([1, 2], [10, 15], color=improved_colors)
        ax[1].set_xticks([1, 2])
        ax[1].set_xticklabels(["Blue", "White"])
        ax[1].set_title("Improved: High Luminance Contrast")

    for a in ax:
        a.set_ylim(0, 20)
        a.set_ylabel("Value")
    plt.tight_layout()
    plt.show()

# Interactive widget for challenge types
confusion_selector = widgets.Dropdown(
    options=["Red-Green", "Low-Luminance"],
    description="Challenge Type:"
)

# Display interactive output
interactive_plot = widgets.interactive_output(
    demonstrate_challenges, {"confusion_type": confusion_selector}
)

# Arrange widgets and output
display(widgets.VBox([confusion_selector, interactive_plot]))
````

````python
def plot_solution(hue, luminance, saturation):
    # Generate colors with hue, luminance, and saturation
    colors = []
    for i in range(5):
        h = (hue + i * 30) % 360 / 360
        l = min(max(luminance + (i - 2) * 0.1, 0), 1)
        s = min(max(saturation + (i - 2) * 0.1, 0), 1)
        colors.append((h, s, l))

    # Create visualization
    fig, ax = plt.subplots(1, 5, figsize=(15, 3))
    for i in range(5):
        ax[i].imshow([[colors[i]]], cmap="hsv")
        ax[i].axis("off")
    plt.tight_layout()
    plt.show()

# Sliders for hue, luminance, and saturation
hue_slider = widgets.IntSlider(value=0, min=0, max=360, description="Hue")
luminance_slider = widgets.FloatSlider(value=0.5, min=0, max=1, description="Luminance")
saturation_slider = widgets.FloatSlider(value=0.5, min=0, max=1, description="Saturation")

# Display the interactive tool
display(widgets.VBox([hue_slider, luminance_slider, saturation_slider]))
widgets.interactive_output(plot_solution, {"hue": hue_slider, "luminance": luminance_slider, "saturation": saturation_slider})
````
